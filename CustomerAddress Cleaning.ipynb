2.1 State

```python
CustomerAddress['state'].value_counts()
```

```python
def replace_state_names(state_name):
    if state_name=='New South Wales':
        return 'NSW'
    elif state_name=='Victoria':
        return 'VIC'
    else :
        return state_name
CustomerAddress['state'] = CustomerAddress['state'].apply(lambda x: replace_state_names(x))
```
